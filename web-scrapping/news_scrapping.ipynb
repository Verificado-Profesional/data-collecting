# GoogleNews

```python
from pygooglenews import GoogleNews
import json
import time
import pandas as pd
```

```python
def google_news_to_df(noticia):
    df_row = {}
    
    df_row['Titulo'] = noticia['title']
    df_row['Contenido'] = None
    df_row['Fecha'] = noticia['published']
    df_row['Fuente'] = noticia['source']['href']
    df_row['Categoría'] = None
    df_row['Autores'] = None
    df_row['Link'] = noticia['link']
    df_row['Clasificación'] = None
    df_row['Narrativa'] = None


    return df_row
```

```python
def top_news_dataframe(google_news: GoogleNews):
    news = google_news.top_news()
    news_df = pd.DataFrame(columns=['Titulo', 'Contenido', 'Fecha', 'Fuente', 'Categoría', 'Autores', 'Link', 'Clasificación', 'Narrativa'])

    for noticia in news['entries']:
        news_df = pd.concat([news_df, pd.DataFrame(google_news_to_df(noticia), index=[0])], ignore_index=True)

   #map(lambda noticia: pd.concat([news_df, pd.DataFrame(google_news_to_df(noticia), index=[0])], ignore_index=True), news['entries'])

    return news_df
```

```python
gn = GoogleNews(lang='es-AR', country='AR')
news_df = top_news_dataframe(gn)
```

# NewsCatcher
## Encuentra los articulos de una pagina de noticias

```python
from newscatcher import Newscatcher, describe_url
```

```python
def newscatcher_to_df(noticia, website):
    df_row = {}
    
    df_row['Titulo'] = noticia['title']
    df_row['Contenido'] = noticia['summary']
    df_row['Fecha'] = noticia['published']
    df_row['Fuente'] = website
    df_row['Categoría'] = None
    df_row['Autores'] = None
    df_row['Link'] = noticia['link']
    df_row['Clasificación'] = None
    df_row['Narrativa'] = None


    return df_row
```

```python
def top_news_dataframe(website: str):
    website_news = Newscatcher(website).get_news()['articles'] #habria que eventualmente poner un try catch
    news_df = pd.DataFrame(columns=['Titulo', 'Contenido', 'Fecha', 'Fuente', 'Categoría', 'Autores', 'Link', 'Clasificación', 'Narrativa'])

    for noticia in website_news[:100]: 
        news_df = pd.concat([news_df, pd.DataFrame(newscatcher_to_df(noticia, website), index=[0])], ignore_index=True)

    return news_df
```

```python
top_news_website = top_news_dataframe('infobae.com')
```

# Newspaper3k

```python
import newspaper
import json
```

```python
def buscar_contenido(link):
    article = newspaper.Article(url=link, language='es')
    try:
        article.download()
    except:
        return None
    article.parse()

    return str(article.text)
```

```python
def completar_contenido(news_df):
    for index, row in news_df.iterrows():
        row['Contenido'] = buscar_contenido(row['Link'])

    return news_df
```
